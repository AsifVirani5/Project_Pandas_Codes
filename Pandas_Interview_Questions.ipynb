### Python_Pandas_Indexing_Chunking_Conditions

```python
# Read the excel file and set 2nd column as index

df = pd.read_excel("sales.xlsx", index_col = 1) # Read the excel file with index_col set to 1
```

```python
df
```

```python
# Display sales columns
df["Sales"] # Display the Sales column using column name
```

```python
df.Sales # Display the Sales column using dot operator
```

```python
type(df["Sales"]) # Display the type of the column using column name
```

```python
df[["Profit", "Sales"]].head() # Display the Profit and Sales columns using column names under two square brackets the first square bracket is for the list of columns and the second square bracket is for the dataframe and the second square bracket is for the list of columns
```

```python
 # Display the Southern Asia row using loc function
df.loc["Southern Asia", :] # Display the Southern Asia row using loc function
```

```python
df.loc["Southern Asia", "Sales"] # Display the Southern Asia row and Sales column using loc function
```

```python
df.iloc[6] # Display the 6th row using iloc function
```

```python
df.iloc[6, 3] # Display the 6th row and 3rd column using iloc function
```

```python
df.loc[:, ["Market", "Profit", "Sales"]] # Display the Market, Profit and Sales columns using loc function and ":" denotes all rows
```

```python
df.iloc[:, [0, 2, 3]] # Display the 1st, 2nd and 3rd columns using iloc function and ":" denotes all rows
```

```python
df.iloc[[0,2,3], :] # Display the 1st, 3rd and 4th rows using iloc function and ":" denotes all columns
```

```python
df.loc[["Western Africa", "Southern Africa"], :] #  Display the Western Africa and Southern Africa rows using loc function and ":" denotes all columns
```

```python
df.loc[["Western Africa","Southern Africa"], ["Profit", "Sales"]] # Display the Western Africa and Southern Africa rows and Profit and Sales columns using loc function
```

```python
df[df["Sales"]> 300000] # Display the rows where Sales is greater than 300000
```

```python
df[ (df["Market"].isin(["LATAM", "Europe"])) & (df["Sales"]> 250000) ] # Display the rows where Market is LATAM or Europe and Sales is greater than 250000
```
